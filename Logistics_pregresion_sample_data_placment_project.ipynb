```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#load data
Data = pd.read_csv('/content/placement.csv')
Data
```

```python
#cleaning the data
Data.isnull().sum()
```

```python
#dividing data into input output
X = Data[['cgpa','placement_exam_marks']]
X
```

```python
y = Data[['placed']]
y
```

```python
#we will perfom train_test_split
from sklearn.model_selection import train_test_split
```

```python
#now we will train the data
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
#we will now creat the model
from sklearn.linear_model import LogisticRegression
Model = LogisticRegression()
```

```python
#now we will fit the data in model
Model.fit(X_train,y_train)
```

```python
#now we will predict the values
y_pred = Model.predict(X_test)
y_pred
```

```python
#we will checking the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```

```python
#our model is 51.50% accurate
```
